## 4. Collocations and bigrams

```python
# a collocation is a sequence of words that occur together unusually often
# to get a handle on collocations, we start off by extracting from a text a list of word pairs, also known as bigrams
list(bigrams(['more', 'is', 'said', 'than', 'done']))
```

```python
# Now, collocations are essentially just frequent bigrams, except that we want to pay more attention to the cases that involve rare words. In particular, we want to find bigrams that occur more often than we would expect based on the frequency of the individual words.
# they are fixing this bug as of October 2019
text4.collocations()
```
